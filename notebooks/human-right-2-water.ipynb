```python
import os.path as op
import pandas as pd
import geopandas as gpd
from cawaterchallenge.utils import get_unique_per_source
```

```python
paths = ['/home/robert/data/ca-water-challenge/cehtp/water-system-service-areas-current-complete-converted.geojson',
         '/home/robert/data/ca-water-challenge/cehtp/water-system-service-areas-not-current-complete-converted.geojson',
         '/home/robert/data/ca-water-challenge/cehtp/water-system-service-areas-current-not-complete-converted.geojson',
         '/home/robert/data/ca-water-challenge/cehtp/water-system-service-areas-not-current-not-complete-converted.geojson'
        ]

ce = []
for path in paths:
    df = gpd.read_file(path)
    ce.append(df)
ce = pd.concat(ce)

ce.rename(columns={'pwsid': 'water_system_id'}, inplace=True)
```

```python
hr = []
paths = ['/home/robert/data/ca-water-challenge/hr2w/hr2w_web_data_active_06-2018.xlsx',
         '/home/robert/data/ca-water-challenge/hr2w/hr2w_web_data_rtc_06-2018.xlsx']
for path in paths:
    df = pd.read_excel(path)
    print(path, len(df))
    df['source_file'] = op.basename(path)
    hr.append(df)

hr = pd.concat(hr)

columns = {k: k.lower() for k in hr.columns}
hr.rename(columns=columns, inplace=True)
hr.rename(columns={'water_system_number': 'water_system_id'}, inplace=True)
```

```python
get_unique_per_source([ce, hr], 'water_system_id', ['ce', 'hr'])
```
